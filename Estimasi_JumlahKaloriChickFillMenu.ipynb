# Deskripsi library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df = pd.read_csv('chick-fil-a-nutrition.csv')
```

# Deskripsi Dataset

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah Kalori Berdasarkan Serving Size

```python
models = df.groupby('Serving size').count()[['Protein(G)']].sort_values(by='Protein(G)',ascending=True).reset_index()
models = models.rename(columns={'Protein(G)':'KaloryServingSize'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Serving size'], y=models['KaloryServingSize'], color='royalblue')
plt.xticks(rotation=60)
```

# Distribusi Jumlah Kalori

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Calories'])
```

# Seleksi Fitur

```python
features = ['Protein(G)','Fat(G)','Cholesterol(MG)','Carbohydrates(G)','Sugar(G)','Sodium(MG)']
x = df[features]
y = df['Calories']
x.shape, y.shape
```

# Split Data Training Dan Data Testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat Model Regresi Linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# Membuat Inputan Model Regresi Linier

```python
#Protein : 34 , Fat (G) : 40.0 , Cholesterol (MG) : 415 , Carbohydrates (G) : 51 , Sugar (G)  : 2 , Sodium (MG) : 1750	
input_data = np.array([[ 34,40.0,415,51,2,1750]])

prediction = lr.predict(input_data)
print('Estimasi Jumlah Kalori Menu Chick-fil-A:', prediction, 'kCal')
```
